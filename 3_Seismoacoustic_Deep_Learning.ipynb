## Comparing Models
---
#### Loading results

```python
%%time
# Load seismic histories
split = 0
histories_seismic = []
while split < 5:
    history_tmp = np.load('/Volumes/Extreme SSD/Korea_Events/Models/Seismic/history_split'+str(split+1)+'.npy',allow_pickle='TRUE').item()
    histories_seismic.append(history_tmp)
    split += 1
# Calculate average and max for seismic
max_scores_seismic = np.zeros((len(histories_seismic)))
avg_scores_seismic = np.zeros((len(histories_seismic)))
std_scores_seismic = np.zeros((len(histories_seismic)))
max_score_loss_seismic = np.zeros((len(histories_seismic)))
avg_loss_seismic = np.zeros((len(histories_seismic)))
std_loss_seismic = np.zeros((len(histories_seismic)))
for i in range(len(histories_seismic)):
    max_scores_seismic[i] = np.max(histories_seismic[i]['val_f1_score'][-11:])
    avg_scores_seismic[i] = np.mean(histories_seismic[i]['val_f1_score'][-11:])
    std_scores_seismic[i] = np.std(histories_seismic[i]['val_f1_score'][-11:])
    max_score_loss_seismic[i] = histories_seismic[i]['val_loss'][(np.where((histories_seismic[i]['val_f1_score'] == np.max(histories_seismic[i]['val_f1_score'][-11:])))[0]).min()]
    avg_loss_seismic[i] = np.mean(histories_seismic[i]['val_loss'][-11:])
    std_loss_seismic[i] = np.std(histories_seismic[i]['val_loss'][-11:])
#-----------------------------------------------------------------------------------------------------------------------#
# Load seismoacoustic histories
split = 0
histories_seismoacoustic = []
while split < 5:
    history_tmp = np.load('/Volumes/Extreme SSD/Korea_Events/Models/Seismoacoustic/history_split'+str(split+1)+'.npy',allow_pickle='TRUE').item()
    histories_seismoacoustic.append(history_tmp)
    split += 1
# Calculate average and max for seismoacoustic
max_scores_seismoacoustic = np.zeros((len(histories_seismoacoustic)))
avg_scores_seismoacoustic = np.zeros((len(histories_seismoacoustic)))
std_scores_seismoacoustic = np.zeros((len(histories_seismoacoustic)))
max_score_loss_seismoacoustic = np.zeros((len(histories_seismoacoustic)))
avg_loss_seismoacoustic = np.zeros((len(histories_seismoacoustic)))
std_loss_seismoacoustic = np.zeros((len(histories_seismoacoustic)))
for i in range(len(histories_seismic)):
    max_scores_seismoacoustic[i] = np.max(histories_seismoacoustic[i]['val_f1_score'][-11:])
    avg_scores_seismoacoustic[i] = np.mean(histories_seismoacoustic[i]['val_f1_score'][-11:])
    std_scores_seismoacoustic[i] = np.std(histories_seismoacoustic[i]['val_f1_score'][-11:])
    max_score_loss_seismoacoustic[i] = histories_seismoacoustic[i]['val_loss'][(np.where((histories_seismoacoustic[i]['val_f1_score'] == np.max(histories_seismoacoustic[i]['val_f1_score'][-11:])))[0]).min()]
    avg_loss_seismoacoustic[i] = np.mean(histories_seismoacoustic[i]['val_loss'][-11:])
    std_loss_seismoacoustic[i] = np.std(histories_seismoacoustic[i]['val_loss'][-11:])
```
